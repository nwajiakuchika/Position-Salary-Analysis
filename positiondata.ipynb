```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Reading of file

```python
df = pd.read_csv('position_salaries.csv')
df
```

```python
X = df[['Level']]
Y = df['Salary']
```

```python
X
```

```python
Y
```

# MODEL FIT

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

# MODEL FIT FOR POLYNOMIAL REGRESSION

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures (degree = 5)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,Y)
```

# PREDICTING THE FEATURE TEST RESULT

```python
X_predict = lin_reg.predict(X)
X_predict
```

# PREDICTING THE FEATURES RESULT FOR POLYNOMIAL REGRESSION ⬇️

```python
Y_lin_predict = lin_reg_2.predict(X_poly)
Y_lin_predict
```

# VISUALIZE POLYNOMIAL LINEAR REGRESSION ⬇️

```python
plt.scatter(X,Y, color='red')
plt.plot(X,Y_lin_predict,color = 'blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show
```

# TO MAKE THE REGRESSION LINE A PERFECT SHAPE ⬇️

```python
X_rangeP = np.linspace(X.min(),X.max(),100).reshape(-1,1)
X_range_P = poly_reg.transform(X_rangeP)
Y_range_predP = lin_reg_2.predict(X_range_P)

plt.scatter(X,Y, color = 'red')
plt.plot(X_rangeP,Y_range_predP,color='black')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

# SHOWING IT IN TABULAR FORM ⬇️

```python
df_results = pd.DataFrame({
    "Level": X["Level"],
    "Actual Salary":Y,
    "Predicted Salary":Y_lin_predict.astype(int)
})
df_results
```
